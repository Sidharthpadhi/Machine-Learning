## Hyperparameter Optimization

#### Lets say you come up with some features after spending so much time on feature engineering and aslo select a model that best fit the problem statement.

#### Now instead of doing hyperparameter tuning manually we can use different techniques to select best hyperparameters without investing a lot of time.

#### Hyperparameter optimization can be done in many different ways.

#### Lets say we can use it to find best learning rate or in neural network i.e no. of hidden layers etc.

#### Everything we have to do inside cross-validation loop as long as you do that you are not over-fitting the model.

## Grid Search

```python
import pandas as pd
import numpy as np

from sklearn import ensemble
from sklearn import metrics
from sklearn import model_selection

from sklearn import preprocessing
from sklearn import decomposition
from sklearn import pipeline
```

```python
if __name__ == '__main__':
    df = pd.read_csv('train.csv')
    print(df.shape)
    X = df.drop('price_range',axis=1).values
    y = df.price_range.values
    
    classifier = ensemble.RandomForestClassifier(n_jobs=-1)
    # Syntax:- name__param
    # Example:- rf__n_estimators
    param_grid = {
    "n_estimators":[100,200,300,400],
    "max_depth":[1,3,5,7],
    "criterion":["gini","entropy"],
    }
# sk-learn provides a way to do grid-search with cross-validation.
   
    model = model_selection.GridSearchCV(
        estimator=classifier,
        param_grid = param_grid,
        scoring="accuracy",
        verbose=10,
        n_jobs=1,
        cv=5,
    )
    
    model.fit(X,y)

    # Grid search tries all prmutation & combination
```

```python
# Print the best_score_
print(model.best_score_)
```

```python
# Print the best parameters
print(model.best_estimator_.get_params())
```
